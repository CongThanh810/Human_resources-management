```python
categorical_cols = ['BusinessUnit', 'EmployeeStatus', 'EmployeeType', 'PayZone', 
                    'EmployeeClassificationType', 'RaceDesc', 'MaritalDesc', 
                    'Training Program Name', 'Training Type', 'Training Outcome']
for col in categorical_cols:
    plt.figure(figsize=(6, 6))
    values = df[col].value_counts() 
    plt.pie(values, labels=values.index, autopct='%.1f%%', shadow=True)
    plt.title(f'Tỷ Lệ {col}')
    plt.show()
    #Tạo biểu đồ bánh cho từng cột phân loại để minh họa phân bố của chúng.
```

```python
# Creating a function that plot the barchart as wells as a pie chart for the categorical column.
def plot_bar_pie(column, dataframe=df, top_n=5):
    """
    Vẽ biểu đồ cột và biểu đồ bánh bên cạnh nhau.
    """
    # Kiểm tra để đảm bảo cột là phân loại.
    if dataframe[column].dtypes == "O":
        dist = dataframe[column].value_counts(normalize=True).head(top_n)
        # Bắt đầu vẽ biểu đồ cột.
        fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(15, 5))
        ax0.bar(dist.index, dist.values)
        ax0.set_xticklabels(rotation=45, labels=dist.index)
        ax0.set_title(f"Biểu Đồ Cột Phân Phối Cho {column}")
        ax0.set_xlabel(column)
        ax0.set_ylabel("Tần Suất")
        # Vẽ biểu đồ bánh
        ax1.pie(dist, labels=dist.index, autopct="%1.2f%%")
        ax1.set_title(f"Biểu Đồ Bánh Phân Phối Cho {column}")
        return fig.show()
    else:
        print("Cột đã cho không phải là cột phân loại")
        
```

```python
plot_bar_pie("Title")
```
